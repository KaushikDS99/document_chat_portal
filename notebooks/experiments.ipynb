## Data Ingestion

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
file_path = "/Users/kaushikchakraborty/Downloads/attention.pdf"
loader = PyPDFLoader(file_path)
```

```python
loader
```

```python
document = loader.load()
```

```python
print(document[0].page_content)
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, 
                                               chunk_overlap=50,
                                               length_function=len)
```

```python
text_chunks = text_splitter.split_documents(document)
```

```python
text_chunks
```

```python
text_chunks[0].metadata
```

```python
text_chunks[0].page_content
```

```python
from langchain_community.vectorstores import FAISS
```

```python
##FAISS- It is a in memory vector store. We can persist over a disk.
vector_store = FAISS.from_documents(text_chunks, embedding_model)
```
